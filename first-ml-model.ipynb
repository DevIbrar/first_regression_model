```python
# Your imports
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
data = pd.read_csv("/kaggle/input/dataset/train_energy_data.csv").dropna()
test = pd.read_csv("/kaggle/input/dataset/test_energy_data.csv").dropna()

# Mapping
building_type_map = {"Residential": 1, "Commercial": 2, "Industrial": 3}
day_of_week_map = {"Weekday": 1, "Weekend": 0}

for df in [data, test]:
    df["Building Type"] = df["Building Type"].map(building_type_map)
    df["Day of Week"] = df["Day of Week"].map(day_of_week_map)

# Train/Test Split
X = data.iloc[:, 0:6]
y = data.iloc[:, 6]

X_test = test.iloc[:, 0:6]
y_test = test.iloc[:, 6]

# Train Model
lr = LinearRegression()
lr.fit(X, y)

# Predict
pred = lr.predict(X_test)

# Evaluation
r2 = r2_score(y_test, pred)
mse = mean_squared_error(y_test, pred)

print(f"R² Score: {r2:.3f}")
print(f"MSE: {mse:.2f}")
```
